```python
#import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline     
```

```python
pl = pd.read_csv('/Users/shub1801/Downloads/EPL_20_21.csv')
pl.head()
```

```python
pl.info()
```

```python
pl.describe()
```

```python
pl['MinsPerMatch'] = (pl['Mins']/pl['Matches']).astype(int)
pl['GoalsPerMatch'] = (pl['Goals']/pl['Matches'])
pl['AssistsPerMatch'] = (pl['Assists']/pl['Matches'])
```

```python
pl.head()
```

```python
#Total Goals

pl['Goals'].sum()
```

```python
#Penalty Goals

a = pl['Penalty_Goals'].sum()
```

```python
#Penalty Attempts

b = pl['Penalty_Attempted'].sum()
```

```python
#Pie chart for penalties missed vs scored

plt.figure(figsize=(10,6))
penalty_missed = b - a
data = [penalty_missed,a]
labels=['Penalties Missed','Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')  #autopct converts string to percentage
plt.show()
```

```python
pl[(pl['Club']== 'Manchester United')].head()
```

```python
mid_mu = pl[(pl['Club']== 'Manchester United')&(pl['Position']=='MF')]
```

```python
mid_mu
```

```python
unitedgoals = pl[(pl['Club']== 'Manchester United')&(pl['Goals']>=1)]
unitedgoals.head()
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(unitedgoals['Name'],unitedgoals['Goals'], color ='maroon',width = 0.3) 
plt.xlabel("Players")
plt.ylabel("Number of goals")
plt.xticks(rotation=45, ha='right')
plt.title("Goals by Manchester United Players in the 2020-2021 Season")
plt.show()
```

```python
#goals by man united midfield players
fig = plt.figure(figsize = (10, 5))
plt.bar(mid_mu['Name'],mid_mu['Goals'], color ='maroon',width = 0.3) 
plt.xlabel("Players")
plt.ylabel("Number of goals")
plt.title("Goals by Manchester United Midfielders in the 2020-2021 Season")
plt.show()
```

```python
#non-penalty goals by man united midfield players
fig = plt.figure(figsize = (10, 5))
goals = (mid_mu['Goals'] - mid_mu['Penalty_Goals']).astype(int)
plt.bar(mid_mu['Name'],goals, color ='navy',width = 0.3) 
plt.xlabel("Players")
plt.ylabel("Number of goals")
plt.title("Goals by Manchester United Midfielders in the 2020-2021 Season")
plt.show()
```

```python
#midfielders who played 2000 or more minutes and scored more than 5 goals
over5 = pl[(pl['Position']=='MF') & (pl['Mins']>=2000) & (pl['Goals']>=5)]
over5
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(over5['Name'],over5['Goals'], color ='green',width = 0.3) 
plt.xlabel("Players")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Number of goals")
plt.title("Goals by Midfielders who played 2000+ minutes and scored more than 5 goals")
plt.show()
```

```python
#midfielders who played 2000 or more minutes and assisted more than five times
ov5 = pl[(pl['Position']=='MF') & (pl['Mins']>=2000) & (pl['Assists']>=5)]
ov5
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(ov5['Name'],ov5['Assists'], color ='brown',width = 0.3) 
plt.xlabel("Players")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Number of Assists")
plt.title("Assists by Midfielders who played 2000+ minutes and assisted more than 5 goals")
plt.show()
```

```python
#non-penalty goals by midfield players by playing over 2000+ mins and scoring 5 or more times
fig = plt.figure(figsize = (10, 5))
goals = (over5['Goals'] - over5['Penalty_Goals']).astype(int)
plt.bar(over5['Name'],goals, color ='navy',width = 0.3) 
plt.xlabel("Players")
plt.ylabel("Number of Goals")
plt.xticks(rotation=45, ha='right')
plt.title("Non-Penalty Goals by midfielders who played 2000+ minutes and scored more than 5 goals in the 2020-2021 Season")
plt.show()
```

```python
totass = pd.DataFrame(pl.groupby('Club',as_index=False)['Assists'].sum())
totass
```

```python
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=totass.sort_values(by="Assists") , palette="Set2")
ax.set_xlabel("Club", fontsize=30)
ax.set_ylabel("Assists", fontsize=20)
plt.xticks(rotation=70)
plt.rcParams["figure.figsize"] = (10,10)
plt.title("Plot of Clubs vs Total Assists ", fontsize=20)
plt.show()
```

```python
# top 10 players making assists
top_10_assists = pl[['Name' , 'Club', 'Assists','Matches']].nlargest(n=10,columns = 'Assists')
top_10_assists
```

```python
Goals_by_clubs = pd.DataFrame(pl.groupby('Club', as_index=False)["Goals"].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=Goals_by_clubs.sort_values(by="Goals") , palette="rocket")
ax.set_xlabel("Club", fontsize=30)
ax.set_ylabel("Goals", fontsize=20)
plt.xticks(rotation=70)
plt.rcParams["figure.figsize"] = (12,6)
plt.title("Plot of Clubs vs Total Goals ", fontsize=20)
plt.show()
```
